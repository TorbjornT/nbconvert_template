```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['savefig.dpi'] = 200
mpl.rcParams['figure.figsize'] = (4,2.4)
mpl.rcParams['savefig.pad_inches'] = 0
```

# First part

And then some fancy words.

```python
x = np.linspace(0, 3*np.pi,500)
y1 = np.sin(x)
y2 = 0.5*np.cos(x) + 0.5
fig, ax = plt.subplots()
ax.plot(x,y1,label='sin')
ax.plot(x,y2,label='cos')
ax.legend(loc='best')
ax.set_xlim((0,10))
ax.set_ylim((-1.1,1.1))
fig.tight_layout()
fig.savefig('demofig.pdf')
```
